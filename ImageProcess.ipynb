```python
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
sample_01 = 'FH_I0000922_1_mask.bmp'
sample_02 = 'FH_I0007729_mask.bmp'
sample_03 = 'FN_I0007519_mask.bmp'
sample_04 = 'FN_I0020375_mask.bmp'
sample_05 = 'FN+1_I0013163_mask.bmp'
```

```python
flist = sorted(os.listdir('/data/sdi/datasets/CPN/Masks'))
img = os.path.join('/data/sdi/datasets/CPN/Masks/', sample_01)

image = Image.open(img).convert("L")
plt.figure(figsize=(10,10))
plt.imshow(image, cmap='gray')
```

```python
src = cv2.imread(img)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
dst = cv2.morphologyEx(src, cv2.MORPH_OPEN, kernel, iterations=9)
plt.figure(figsize=(10,10))
plt.imshow(dst)
```
