```python
src = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 5))
dst = cv2.morphologyEx(src, cv2.MORPH_ERODE, kernel, iterations=1)
dst = cv2.morphologyEx(dst, cv2.MORPH_DILATE, kernel, iterations=1)
plt.figure(figsize=(10,10))
plt.imshow(dst/2, cmap='gray', vmin=0, vmax=255)
```
