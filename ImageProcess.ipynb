```python
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
sample_01 = 'FH_I0000922_1_mask.bmp'
sample_02 = 'FH_I0007729_mask.bmp'
sample_03 = 'FN_I0007519_mask.bmp'
sample_04 = 'FN_I0020375_mask.bmp'
sample_05 = 'FN+1_I0013163_mask.bmp'
```

```python
flist = sorted(os.listdir('/data/sdi/datasets/CPN/Masks'))
img = os.path.join('/data/sdi/datasets/CPN/Masks/', sample_05)

image = Image.open(img).convert("L")
plt.figure(figsize=(10,10))
plt.imshow(image, cmap='gray')
```

```python
src = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 5))
dst = cv2.morphologyEx(src, cv2.MORPH_ERODE, kernel, iterations=1)
dst = cv2.morphologyEx(dst, cv2.MORPH_DILATE, kernel, iterations=1)
plt.figure(figsize=(10,10))
plt.imshow(dst/2, cmap='gray', vmin=0, vmax=255)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(dst/2 + src/2, vmin=0, vmax=255)
```
